***Importing all the Necessary libraries***

```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from skimage import io
from skimage.transform import resize

# Specify the path to your dataset
data_dir = 'C:/Users/manis/Downloads/archive/train/train/'

# Function to load images and labels
def load_data(directory):
    images = []
    labels = []
    class_mapping = {}
    current_label = 0

    for class_folder in os.listdir(directory):
        class_path = os.path.join(directory, class_folder)
        if os.path.isdir(class_path):
            class_mapping[current_label] = class_folder
            for filename in os.listdir(class_path):
                if filename.endswith(('.jpg', '.jpeg', '.png')):
                    img_path = os.path.join(class_path, filename)
                    img = io.imread(img_path)
                    img = resize(img, (100, 100))  # Resize images to a consistent size
                    images.append(img.flatten())  # Flatten the image into a 1D array
                    labels.append(current_label)

            current_label += 1

    return np.array(images), np.array(labels), class_mapping

# Load data
X, y, class_mapping = load_data(data_dir)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=400, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the classifier
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')

```

```python
# Display classification report
class_names = [class_mapping[i] for i in range(len(class_mapping))]
report = classification_report(y_test, y_pred, target_names=class_names)
print(report)
```
